# **3.0 Data Wrangling:**
## **3.1 Dealing with Datatypes:**
- Let's discussed about types of those columns that are doubteable

```python
df["Content_Rating"].unique()
```

- These values seems to be specific, we don't need to change the dtype of this column.
- We can create another column based on these values.

```python
df["Content_Rating"] = df["Content_Rating"].replace("4+", "Toddler")
df["Content_Rating"] = df["Content_Rating"].replace("9+", "Children")
df["Content_Rating"] = df["Content_Rating"].replace("12+", "Teen")
df["Content_Rating"] = df["Content_Rating"].replace("17+", "Young")
df["Content_Rating"] = df["Content_Rating"].replace("Not yet rated", "Everyone")
```

- As in the above code, we replace 4+ to Toddlers, 9+ to Children, 12+ to Teen, 17+ to Young and Not yet rated to Everyone.

```python
# check again the unique values
df["Content_Rating"].unique()
```

- These all values are specific that's why these values are in string and dtype object of this column

```python
df["Free"].unique()
```

- Free columns means those apps which we can freely download from Appsore.
- We can see that the Free column has two unique values, one is True and other is False.
- This sense may be confused in future for analysis, So we can change the column name and its values. 

```python
df['Free'] = np.where(df['Free'] == True, 'Free', 'Paid') 
# change the column name
df.rename(columns={"Free":"Type"}, inplace=True)
```

```python
df["Type"].unique()
```

- Now it is looking good.

## **3.2 Dealing with Missing Values:**
- Let's check the null values from the dataset.

```python
df.shape
```

```python
# percentage of null values 
df.isnull().sum().sort_values(ascending=False)/len(df) * 100
```

```python
# filling the null values of price and size_bytes of their mean
df["Price"] = df.Price.fillna(value= df["Price"].mean())
df["Size_Bytes"] = df.Size_Bytes.fillna(value= df["Size_Bytes"].mean())

```

```python
df.isnull().sum()
```

```python
df["Primary_Genre"].unique()
```

```python
df["Content_Rating"].unique()
```

```python
df["Size_Bytes"].unique()
```

```python
# Now we have to convert the size of bytes to mb
df["Size_Bytes"] = df["Size_Bytes"].apply(lambda x: (x/1024)/1024)
```

```python
print(df["Size_Bytes"].unique())
print(df["Size_Bytes"].dtype)
```

```python
df.rename(columns={"Size_Bytes": "Size_MB"}, inplace=True)
df['Size_MB'] = df["Size_MB"].round().astype(int)
df.head()
```

```python
df["Average_User_Rating"].unique()
```

```python
# Round off the numeric values and change its type to integer
df["Average_User_Rating"] = df["Average_User_Rating"].round().astype(int)
```

```python
df["Current_Version_Score"] = df["Current_Version_Score"].round().astype(int)
df["Current_Version_Score"].unique()
```

# **4.0 Data Visulization:**

```python
most_reviewed_app = df.groupby("Primary_Genre", as_index=False)["Reviews"].sum().sort_values(by = "Reviews", ascending=False).head(10)
print(most_reviewed_app)

fig = plt.subplots(1, figsize=(10,6), facecolor="#fce75d") 

fig = sns.barplot(data=most_reviewed_app, x = "Primary_Genre", y = "Reviews")
plt.xticks(rotation=45)
# add grid lines
fig.grid(True, linestyle="--", linewidth = 0.5)
# set title, xlabel and ylabel
plt.title("Top 10 Apps by Number of Reviews", fontsize=14, color="black")
plt.xlabel("Type of App", fontsize=14, color="black")
plt.ylabel("Reviews", fontsize=14, color="black")
# set background colour of bars
fig.set_facecolor('#FFF7D1') 
```

```python
average_app_size = df.groupby("Primary_Genre", as_index=False)["Size_MB"].mean().sort_values(by = "Size_MB", ascending=False).head(10)
print(average_app_size)

fig = plt.subplots(1, figsize=(10,6), facecolor="#fce75d") 

fig = sns.barplot(data=average_app_size, x = "Primary_Genre", y = "Size_MB")
plt.xticks(rotation=45)
# add grid lines
fig.grid(True, linestyle="--", linewidth=0.5)
# set title, xlabel and ylabel
plt.title("To 10 Apps by Size", fontsize=14, color="black")
plt.xlabel("Type of App", fontsize=14, color="black")
plt.ylabel("Size of App", fontsize=14, color="black")
# set background colour of bars
fig.set_facecolor('#FFF7D1') 

for bars in fig.containers:
    fig.bar_label(bars)
```

```python
# create two figures with 1 row and 2 column
fig, ax = plt.subplots(1, 2, figsize=(10,6), facecolor="#fce75d") 

# create a countplot using seaborn
sns.countplot(data= df, x = "Type", ax = ax[0], palette = [ "#047d69", "#05f7cf"])
# set the title of chart
ax[0].set_title("Counts of Free vs Paid Users", fontsize=14, color="black")
# add grid lines
ax[0].grid(True, linestyle="--", linewidth=0.5)
# set background colour of bars
ax[0].set_facecolor('#FFF7D1') 
# set x-label and y-label
ax[0].set_xlabel("Type", fontsize=14, color="black")
ax[0].set_ylabel("Counts", fontsize = 14, color="black")

# create a piechart using matplotlib
size = df["Type"].value_counts()
labels = size.index

# create a pie chart
plt.pie(size, labels= labels, autopct='%1.1f%%',
        startangle = 140, colors = ["#bd061e", "#f53851"])

# set title of chart
ax[1].set_title("Distribution of Free vs Paid Users", fontsize=14, color="black")

# create a space between plots
plt.subplots_adjust(wspace=0.1)

# print the values counts of Type
print(df["Type"].value_counts())
```

```python
# create two figures with 1 row and 2 column
fig, ax = plt.subplots(1, 2, figsize=(10,6), facecolor="#fce75d") 

# create a first count plot
sns.countplot(data= df, x = "Content_Rating", ax = ax[0], palette = ["#026151", "#03ad91", "#02c9a8", "#09ebc5", "#05f7cf"])
# set grid line in the chart
ax[0].grid(True, linestyle="--", linewidth=0.5)
# set background colour of bars
ax[0].set_facecolor('#FFF7D1') 
# set title of chart
ax[0].set_title("Counts of Content Ratings by age group", fontsize=14, color="black")
# set x-label and y-label
ax[0].set_xlabel("Content Ratings", fontsize=14, color="black")
ax[0].set_ylabel("Counts", fontsize = 14, color="black")

# create a pie chart
app_content_ratings = df["Content_Rating"].value_counts()
plt.pie(app_content_ratings, labels= app_content_ratings.index, autopct='%1.1f%%',
        startangle = 140, colors = ["#910316", "#bd061e", "#e30b28", "#f53851", "#f55d71"])
# create title of chart
ax[1].set_title("Distribution of Content Ratings of Users", fontsize=14, color="black")

# create a space between plots
plt.subplots_adjust(wspace=0.1)

# print value counts of Content_Rating
print(app_content_ratings)
```

```python
fig, ax = plt.subplots(1, figsize=(10,6), facecolor="#fce75d") 

fig = sns.countplot(df, x="Content_Rating", hue="Type", palette = [ "#047d69", "#05f7cf"])
# add grid lines
fig.grid(True, linestyle="--", linewidth = 0.5)
# set title, xlabel and ylabel
plt.title("Counts o Content Rating Generation by Type", fontsize=14, color="black")
plt.xlabel("Types of Content Rating Generation", fontsize=14, color="black")
plt.ylabel("Counts", fontsize=14, color="black")
# set background colour of bars
fig.set_facecolor('#FFF7D1') 

for bars in fig.containers:
    fig.bar_label(bars)
```

> # **Conclusion:**

> ## Free App Dominance: 
A significant majority (91.6%) of apps on the Apple Store are free. This indicates that the freemium model dominates the platform, likely driven by users’ preference for cost-free options.

> ## Toddlers as Major Consumers: 
Toddlers represent the largest demographic in terms of app downloads, accounting for 79.7%. This shows a growing trend of parents downloading apps for educational and entertainment purposes for younger children. Even in the category of paid apps, toddlers remain the primary users, highlighting the importance of this age group for app developers.

> ## Young Generation’s Preference: 
Only 10% of downloads are attributed to the younger generation, indicating that they may prefer alternative app sources or have different consumption habits compared to toddlers.

> ## Most Reviewed App Categories: 
Game apps receive the highest number of reviews, reflecting their popularity and engagement level. Following games, finance apps come in second, likely due to increased interest in personal finance management. Other categories like photo & video, shopping, and food & drink apps also receive substantial engagement in terms of reviews, which showcases the broad usage patterns across different sectors.

> ## App Size Trends: 
Graphic design apps have the largest file sizes, averaging 151 MB, likely due to the intensive resources required for design tools. Games are a close second at 142 MB, reinforcing the idea that complex graphics and interactive elements increase app size
